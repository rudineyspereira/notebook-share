## Conversão de endereços em Coordenadas Latitude e Longitude<br>.
Para isso usamos a biblioteca [GeoPy](https://geopy.readthedocs.io/en/stable/#).<br>

```python
from geopy.geocoders import Nominatim
```

```python
localizador = Nominatim(user_agent="myGeocoder")
```

```python
localizacao = localizador.geocode("Universidade Federal de Santa Maria")
```

```python
# Importar a biblioteca Pandas
import pandas as pd
```

```python
vacina_RS = pd.read_csv("/home/rudiney/Downloads/vacina_RS_20072021.csv",sep=";")
```

```python
vacina_RS.head()
```

```python
# Nome das colunas do dataframe 'vacinas_RS'
vacina_RS.columns
```

```python
#
vacina_RS_paciente = vacina_RS[['paciente_endereco_nmmunicipio','paciente_endereco_uf','paciente_endereco_nmpais']]
```

```python
vacina_RS_paciente
```

```python
# Eliminar informações repetidas e para isso, usa-se o método drop_duplicates()
vacina_RS_paciente.drop_duplicates(inplace=True)
```

```python
# dados normalizados e apenas com valores não repetidos
vacina_RS_paciente
```

```python
# Agora é preciso desnormalizar e criar uma nova coluna que compreende a concatenação das 3 primeiras colunas 
vacina_RS_paciente['paciente_endereco'] = vacina_RS_paciente['paciente_endereco_nmmunicipio']+','+vacina_RS_paciente['paciente_endereco_uf']+','+vacina_RS_paciente['paciente_endereco_nmpais']
```

```python
vacina_RS_paciente
```

```python
# Para fazer a leitura dos endereços (4ª coluna) e fazer a conversão utiliza-se o método RateLimiter
from geopy.extra.rate_limiter import RateLimiter
```

```python
geocode = RateLimiter(localizador.geocode, min_delay_seconds=0.2)
```

```python
vacina_RS_paciente['paciente_endereco_geolocalizado'] = vacina_RS_paciente['paciente_endereco'].apply(geocode)
```

```python
vacina_RS_paciente
```

```python
# Eliminar coluna de dataframe
# del vacina_RS_paciente['vacina_paciente_geolocalizado']
```

```python
vacina_RS_paciente.head()
```

```python
vacina_RS_paciente['paciente_endereco_geolocalizado'][0]
```

```python
# Criando coluna com dados de latitude
vacina_RS_paciente['paciente_endereco_geolocalizado'].apply(lambda loc: loc.latitude if loc else None)
```

```python
# Criando coluna com dados de longitude
vacina_RS_paciente['paciente_endereco_geolocalizado'].apply(lambda loc: loc.longitude if loc else None)
```

```python
vacina_RS_paciente['paciente_endereco_coordenadas'] = vacina_RS_paciente['paciente_endereco_geolocalizado'].apply(lambda loc: tuple(loc.point) if loc else None)
```

```python
vacina_RS_paciente.head()
```

```python
# Criando colunas de latitude e longitude separadas no dataframe 'vacina_RS_paciente'
vacina_RS_paciente['paciente_endereco_latitude'] = vacina_RS_paciente['paciente_endereco_geolocalizado'].apply(lambda loc: loc.latitude if loc else None)
vacina_RS_paciente['paciente_endereco_longitude'] = vacina_RS_paciente['paciente_endereco_geolocalizado'].apply(lambda loc: loc.longitude if loc else None)
```

```python
vacina_RS_paciente.head()
```

```python
# Quantos endereços geolocalização não tiveram dados de coordenadas conhecidas
vacina_RS_paciente.paciente_endereco_geolocalizado.isna().sum()
```
